# 2. Use the variance function from the textbook to find the variance of this list. Assume that the list represent a population in whole. 

```python
def mean(lst):
    n = len(lst)
    mean_ = total(lst) / n
    return mean_
```

```python
def total(lst):
    total_ = 0
    for val in lst:
        total_ += val
    return total_
```

```python
def variance(lst, sample = True):
    list_mean = mean(lst)
    n = len(lst)
    DoF = n - 1
    sum_sq_diff = 0
    for val in lst:
        sum_sq_diff += (val - list_mean) ** 2
    if sample == False:
        variance_ = sum_sq_diff / n
    else:
        variance_ = sum_sq_diff / DoF
    
    return variance_

variance1 = variance(list_random1, sample = False)
print("Variance of list_random1:", variance1)
```

# 6. Explain why it might be advantageous to create a function instead of building all commands from scratch as you create a script.

It is advantageous to create a function instead of building all commands from scratch as you create a script because it saves time. Once you have the function made, you no longer have to go back and type the same thing over and over again. You can just quickly use the funtion. It is also advantageous because it will save you from making unnecessary mistakes in your coding.
