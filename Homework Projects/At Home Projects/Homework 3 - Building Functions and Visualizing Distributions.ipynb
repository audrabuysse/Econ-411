# 7. Create a function that plots 2 to 3 different variables from the dataframe. The function may look something like "def plot_data(df,names, etc...). Each plot should include a legend and title.

```python
import matplotlib.pyplot as plt

num_vars = 3
fig, axs = plt.subplots(1,3, figsize = (20,6))

for i in range(num_vars):
    ax = axs[i]
    key = list(dataset.keys())[i+2]
    color = "C" + str(i)

    dataset[key].plot.line(color = color, ax = ax, label = key)
    ax.set_title(key,color = "000000", fontsize = 20)
   
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt

num_vars = 3
fig, axs = plt.subplots(1,3, figsize = (20,6))

for i in range(num_vars):
    ax = axs[i]
    key = list(dataset.keys())[i+2]
    color = "C" + str(i)


    dataset[key].plot.line(color = color, ax = ax, label = key, linewidth = 3.0, linestyle = ':', marker = 'D')
    ax.set_title(key,color = "000000", fontsize = 20)
    ax.set_title((key) + " (Pre-Tax National Income) of Top 10% by State", fontsize = 12)
    ax.set_xlabel("States", fontsize = 17)
    ax.set_ylabel("Pre-Tax National Income", fontsize = 10)

    
```
