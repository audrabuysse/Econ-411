```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from matplotlib.backends.backend_pdf import PdfPages

filename = "Combined Data PP.xlsx"
data = pd.read_excel(filename, header = [4],
                    index_col = [3,1])
data
```

```python
fig, ax = plt.subplots()
data.plot(
    use_index=True, 
    kind='bar',
    x = 'Countries',
    y='Top 1%', 
    ax=ax, 
    color='C3',
    figsize = (15,6)
)
ax1 = ax
ax1.set_xlabel('Countries'),
ax1.set_ylabel('Pre-Tax National Income Share'),
ax.set_title('Top 1% Compared to the Economic Freedom Summary Index in 2020')


data.plot(
    use_index=True, 
    x = 'Countries',
    y='Economic Freedom Summary Index', 
    ax=ax, 
    secondary_y=True, 
    color='blue'
)
ax2 = ax1.twinx()
ax2.set_ylabel('Economic Freedom Summary Index Score'),
ax2.set_xticklabels(ax2.get_xticklabels())
plt.show()
```

```python
fig, ax = plt.subplots()
data.plot(
    use_index=True, 
    kind='bar',
    x = 'Countries',
    y ='Bottom 50%',
    ax = ax,
    color='orange',
    figsize = (15,6)
)
ax1 = ax
ax1.set_xlabel('Countries'),
ax1.set_ylabel('Pre-Tax National Income Share'),
ax.set_title('Bottom 50% Compared to the Economic Freedom Summary Index in 2020')

data.plot(
    use_index=True, 
    x = 'Countries',
    y='Economic Freedom Summary Index', 
    ax=ax, 
    secondary_y=True, 
    color='blue'
)
ax2 = ax1.twinx()
ax2.set_ylabel('Economic Freedom Summary Index Score'),
ax2.set_xticklabels(ax2.get_xticklabels())
plt.show()
```

```python
data.plot(y = 'Top 1%', figsize = (15,6))
plt.xlabel('Countries'),
plt.ylabel('Pre-Tax National Income Share'),
plt.title('Share of Top 1% per Country in 2020')
plt.show()
```

```python
data.plot(y = 'Bottom 50%', figsize = (15,6))
plt.xlabel('Countries'),
plt.ylabel('Pre-Tax National Income Share'),
plt.title('Share of Bottom 50% per Country in 2020')
plt.show()
```

```python
data.plot(y = 'Economic Freedom Summary Index', figsize = (15,6))
plt.xlabel('Countries'),
plt.ylabel('Economic Freedom Summary Index Score'),
plt.title('Economic Freedom Summary Index by Country in 2020')
plt.show()
```
