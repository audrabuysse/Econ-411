### Introduction
The distribution of wealth in a country depends on a variety of factors. Some factors include “legal regulations, taxes, welfare, innovations, macroeconomic conditions, international relations, … and many others” (Burda et al., 2019). The Fraser Institute has been scoring countries based on economic freedom for about fifty years. They score on a range of zero to ten, ten being the highest economic freedom score. The Fraser Institute gives scores on five different factors. These include the size of government, legal systems and property rights, sound money, freedom to trade internationally, and regulation. According to Bergh and Bjørnskov, the size of government is measured “using indicators such as public consumption and transfers relative to the GDP, top marginal tax rates, and state-owned enterprises” (2021). Legal systems and property rights are measured by checking the quality and trustworthiness of legal systems and property rights. The sound money score “captures the effect of high and unpredictable changes in inflation and money supply” (Bergh & Bjørnskov, 2021). The freedom to trade internationally score is based on measures of trade taxes, tariffs, trade barriers, and market controls. Finally, the regulation score is based on “three indices on the regulation of credit, labor, and business” (Bergh & Bjørnskov, 2021). The Fraser Institute also provides an overall economic freedom score (Economic Freedom of the World, 2022). Apergis and Cooray describe, “economic freedom is a multifaceted concept” (2015). While this may be true, I believe that the Fraser Institute did a great job of figuring out a way to measure it. It was no small task to be able to quantify economic freedom, but they made it work. The World Inequality Database (WID) provides data on wealth and income inequality in each country over the years. In terms of wealth inequality, they provide data for the share of the wealth that the top one percent, the top ten percent, and the bottom fifty percent hold (World Inequality Database, 2022). I chose to look at wealth inequality rather than income inequality in this model. I did this because “wealth inequality is even more shocking than income inequality because wealth can generate its own income regarding rent, capital gain, interest, dividend, and can be passed on between generations and thereby most wealth accumulation leads to higher concentration of capital” (Islam, 2018). I have collected data from both the Fraser Institute and the World Inequality Database to use in this project.

### The Model
The data starts in 1995 and ends in 2020, with data collected every 5 years in between. So, I ultimately chose to look at the past 25 years, as I had the full data for them. The first step of this model was to combine the data collected from the Fraser Institute and the World Inequality Database into an Excel spreadsheet. After this, I felt it was necessary to make scatterplots depicting the economic freedom index summary score and its five subcomponents compared to the share of the wealth that the top 1%, top 10%, and bottom 50% held in countries over the 25 years. Each subcomponent was individually compared to the share of wealth held in those 25 years because I thought it would be interesting to see what effect they each had. I made the scatterplots using Matplotlib’s plt.scatter() function.

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

filename1 = "Countries with Full Data.xlsx"
alldata = pd.read_excel(filename1, header = [0],
                    index_col = [])
alldata
```

### The Model
I then decided to run an Ordinary Least Squares (OLS) regression on my data. I thought that this test would help me to describe the relationships that the share that the top 1%, top 10%, and bottom 50% hold have with the economic freedom index and its subcomponents. For this part of my project, I made a new Excel file that included only the countries that had all of their data. This was made up of 98 different countries. After importing this data into Jupyter, I then made three new columns: “Top 1% Change,” “Top 10% Change,” and “Bottom 50% Change.” These columns were made by using Groupby to find the difference between each five-year period’s share of wealth held. I also made new columns and used Groupby to find the difference between each of the economic freedom index categories. I made an overall “Economic Freedom Summary Index Change,” “Size of Government Change,” “Legal System & Property Rights Change,” “Sound Money Change,” “Freedom to trade Internationally Change,” and “Regulation Change.” In using Groupby, I grouped them by their country name, labeled as “Countries,” and their original title name. Since the data I collected started in 1995, it was unable to show the difference between numbers from 1990 to 1995. I had to then replace this space with the mean value in order to get the regression to run. I first started by looking at the change in the share that the top 1% hold. I set the top 1% change as the y-variable and the five components of economic freedom change and the economic freedom index change as the x-variables. I added the x-variables with the y-variable and made the regular variables. I used .corr() to find the pairwise correlation of these regular variables. I added a predictor for the top 1% change. I also added a constant of 1. I then calculated the residuals as well as the SSE, SSR, and SST. After doing all of this, I ran an OLS regression on the data. I also added a graph of the residuals. After completing the OLS regression for the top 1% change, I next reran the same steps but used the top 10% change instead. # Top 1% Regression

### The Results
The results of the regressions probably mean the most to this model. I first conducted the regression for the top 1% change compared to the economic freedom index changes of the 25-year range. The sum of squared explained is about 0.012. The sum of squared residuals is about 0.392. Finally, the sum of squared totals is about 0.404. The estimator variance is about 0.00067. This regression has an r-squared of 0.029 and an adjusted r-squared of 0.019. It has an f-statistic of 2.911 and a log-likelihood of 1315.8. When looking at the bar graph of the residuals, it appears to be fairly normally distributed, with a slight skew to the left. 

```python
y_var = ["Top 1% Change"]
x_vars = ["Size of Government Change", 
          "Legal System & Property Rights Change", 
          "Sound Money Change",
          "Freedom to trade Internationally Change",
          "Regulation Change",
         "Economic Freedom Summary Index Change"]
reg_vars = y_var + x_vars
reg_data = alldata[reg_vars]
reg_data.corr().round(2)
```

```python
reg_data['Top 1% Change'] = reg_data['Top 1% Change'].fillna(reg_data["Top 1% Change"].mean())
reg_data['Size of Government Change'] = reg_data['Size of Government Change'].fillna(reg_data["Size of Government Change"].mean())
reg_data['Legal System & Property Rights Change'] = reg_data['Legal System & Property Rights Change'].fillna(reg_data["Legal System & Property Rights Change"].mean())
reg_data['Sound Money Change'] = reg_data['Sound Money Change'].fillna(reg_data["Sound Money Change"].mean())
reg_data['Freedom to trade Internationally Change'] = reg_data['Freedom to trade Internationally Change'].fillna(reg_data["Freedom to trade Internationally Change"].mean())
reg_data['Regulation Change'] = reg_data['Regulation Change'].fillna(reg_data["Regulation Change"].mean())
reg_data["Economic Freedom Summary Index Change"] = reg_data['Economic Freedom Summary Index Change'].fillna(reg_data["Economic Freedom Summary Index Change"].mean())

reg_data
```

```python
y = reg_data[y_var]
X = reg_data[x_vars]
X["Constant"] = 1
results = sm.OLS(y, X).fit()
```

```python
predictor = results.predict()
reg_data[y_var[0] + " Predictor"] = predictor
reg_data
```

```python
y_hat = reg_data[y_var[0] + " Predictor"]
y_mean = reg_data[y_var[0]].mean()
y = reg_data[y_var[0]]
reg_data["Residuals"] = (y.sub(y_hat))
reg_data["Squared Explained"] = y_hat.sub(y_mean) ** 2
reg_data["Squared Residuals"] = (y.sub(y_hat)) ** 2
reg_data["Squared Totals"] = (y.sub(y_mean)) ** 2
reg_data
```

```python
SSR = reg_data["Squared Explained"].sum()
SSE = reg_data["Squared Residuals"].sum()
SST = reg_data["Squared Totals"].sum()
SSR,SSE,SST
```

```python
n = results.nobs
k = len(results.params)
estimator_variance = SSE / (n - k)
n, k, estimator_variance
```

```python
cov_matrix = results.cov_params()
cov_matrix
```

```python
results.params
```

```python
parameters = {}
for x_var in x_vars:
    parameters[x_var] = {}
    parameters[x_var]["Beta"] = results.params[x_var]
    parameters[x_var]["Standard Error"] = cov_matrix.loc[x_var, x_var] ** (1/2)
    parameters[x_var]["t-stats"] = parameters[x_var]["Beta"] / parameters[x_var]["Standard Error"] 
parameters = pd.DataFrame(parameters).T
parameters
```

```python
r2 = SSR/SST
r2
```

```python
r2_adjusted = 1 - (SSE / (n - k)) / (SST / (n - 1))
r2_adjusted
```

```python
results.summary()
```

```python
reg_data["Residuals"] = results.resid
fig, ax = plt.subplots(figsize = (12,8))

reg_data[["Residuals"]].plot.hist(bins = 100, ax = ax)
```
